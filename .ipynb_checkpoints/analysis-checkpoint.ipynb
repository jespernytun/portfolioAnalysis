## First step: Calculating the beta of a single stock
In this part we'll be caluculating the beta of a single stock (ASML) <br>
For now (or maybe for the rest of the project) we'll be using SPY as a proxy for the rest of the market <br>

This will help me get used to the yahoofinance library, and serve as a guide for later

```python
#importing nessecary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Definition of tickers
ASML = yf.Ticker("ASML") # We want to compare ASML to the market
SPY = yf.Ticker("SPY") # We use the SPY500 as a proxy for the entire market

# Getting historical price data for the last twelwe months, just closing data,
# We translate it to pctchange to facilitate the follwing calculations
# We drop NA values (for some reason the first values are always NA)
ASML_Hist = ASML.history(period="12mo")["Close"].pct_change().dropna()
SPY_Hist = SPY.history(period="12mo")["Close"].pct_change().dropna()
```

```python
# We calculate Cov(ASML, SPY)
cov = ASML_Hist.cov(SPY_Hist)
# We calculate the Var(SPY)
var = SPY_Hist.var()
# beta is defined as cov(R1,R2)/Var(R2), thus we calculate it using the old values
beta = cov/var
print(beta)
```

```python
# Lets trace ASML and SPY for fun

# We make sure that all data points correspond and drop all missing points
df = pd.concat([SPY_Hist.rename("SPY"), ASML_Hist.rename("ASML")], axis=1).dropna()

# We use a scatterplot this time
plt.scatter(df["SPY"], df["ASML"], s=12, alpha=0.6)

# We draw the a line following the beta and one that follows y=x
plt.axline((0, 0), slope=beta, label="slope y=beta")
plt.axline((0, 0), slope=1, color="red", linestyle="--", label="slope y=x")
plt.xlabel("SPY pctchange daily")
plt.ylabel("ASML pctchange daily")
plt.title("ASML vs SPY last 12 months")
plt.legend()
plt.show()
```

## Second step
Now that I calculated the beta of a single stock, we'll continue <br>
This next part is all about defining a function that takes a list of tickers as and calculates the average beta of the entire portfolio

We'll worry about weighting later

```python
# We define the stocks we want to use
MyStocks = ["ASML", "NVO", "EQNR"]
```

```python
def calculateAverageBeta (Stocks):
    
    MarketProxy = yf.Ticker("SPY") # We still use SPY as a proxy for the market in it's entirety
    StockHistory = [] # Here we will store pctchange values for all our Stocks

    # We use yahoofinance to gather all the closing prices the last 12mo
    for i in range(len(Stocks)):
        tempTick = yf.Ticker(Stocks[i])
        tempHist = tempTick.history(period="60mo")["Close"].pct_change().dropna()
        StockHistory.append(tempHist)
    # We gather the same data for MarketProxy
    ProxyHist = MarketProxy.history(period="60mo")["Close"].pct_change().dropna()
        
    # We have now created a list called stock history containing price changes 
    # day by day for our "entire portfolio"

    # We calculate the beta for each and every stock
    BetaList = []
    for data in StockHistory:
        tempBeta = data.cov(ProxyHist) / ProxyHist.var()
        BetaList.append(tempBeta)

    Beta = sum(BetaList) / len(BetaList)
    
    return Beta
```

```python
Beta = calculateAverageBeta(MyStocks)
print(Beta)
# Result 0.95 totally normal, since I have NVO and EQNR in there
```
